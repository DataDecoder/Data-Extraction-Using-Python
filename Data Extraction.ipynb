The Beautiful Soup bundle is utilized to parse the html, that is, take the crude html text and break it into Python objects. 

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# Import url
url = "https://books.toscrape.com/"
html = urlopen(url)
```

```python
# Using the Soup Library
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Scrape the Industries
Title = soup.title
print(Title)
```

```python
# Text Print
text_book = soup.get_text()
```

```python
#Find all the text
soup.find_all('h3')
```

```python
#Remove HTML tags
text_cells = str(text_book)
text_clean = BeautifulSoup(text_cells, "lxml").get_text()
print(text_clean)
```
